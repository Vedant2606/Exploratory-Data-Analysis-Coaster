```python
6. Will try to give some answers based on the analysis.
What are the locations with fastest roller coaster?
```

```python
df['Location']
```

```python
#ignore the location='Other'
ax  = df.query('Location!="Other"')
ax
```

```python
ax  = df.query('Location!="Other"')\
        .groupby('Location')['Speed_mph']\
        .agg(['mean', 'count'])
ax

#this command gives you the count of coaster for particular locations and mean of speed for the coaster on that location
```

```python
ax  = df.query('Location!="Other"')\
        .groupby('Location')['Speed_mph']\
        .agg(['mean', 'count'])\
        .query('count>=10')
ax

#we filter top 10 out of all results
```

```python
ax  = df.query('Location!="Other"')\
        .groupby('Location')['Speed_mph']\
        .agg(['mean', 'count'])\
        .query('count>=10')\
        .plot(kind='barh', title='Average Coaster Speed by location')
ax
#plot the horizintal bar graph 
```

```python
We can sort the values based on mean (AVG SPEED)
```

```python
ax  = df.query('Location!="Other"')\
        .groupby('Location')['Speed_mph']\
        .agg(['mean', 'count'])\
        .query('count>=10')\
        .sort_values('mean')['mean']\
        .plot(kind='barh', title='Average Coaster Speed by location')
```
